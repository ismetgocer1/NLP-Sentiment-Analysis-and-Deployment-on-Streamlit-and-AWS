## 3. Implement Linear Regression

```python
def train_val(y_train, y_train_pred, y_val, y_eval_pred):

    scores = {"train_set": {"R2": r2_score(y_train, y_train_pred),
                            "mae": mean_absolute_error(y_train, y_train_pred),
                            "mse": mean_squared_error(y_train, y_train_pred),
                            "rmse": np.sqrt(mean_squared_error(y_train, y_train_pred))},

              "val_set": {"R2": r2_score(y_val, y_val_pred),
                          "mae": mean_absolute_error(y_val, y_val_pred),
                          "mse": mean_squared_error(y_val, y_val_pred),
                          "rmse": np.sqrt(mean_squared_error(y_val, y_val_pred))}}

    return pd.DataFrame(scores)
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\n", "mae:", mae, "\n", "mse:", mse, "\n", "rmse:", rmse)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
y_val_pred = lm.predict(X_val)
```

```python
y_train_pred = lm.predict(X_train)
```

```python
train_val(y_train, y_train_pred, y_val, y_val_pred)
```

```python
model = LinearRegression()
scores = cross_validate(model, X_train, y_train, scoring=['r2',
                                                          'neg_mean_absolute_error', 'neg_mean_squared_error', 'neg_root_mean_squared_error'], cv=10)
```

```python
pd.DataFrame(scores)
```

```python
pd.DataFrame(scores).iloc[:, 2:].mean()
# If there is inconsistency between the Cross Validation results, there is a problem with the model or data.
```

```python
y_test_pred = lm.predict(X_test)
```

```python
eval_metrics(y_test, y_test_pred)
```

```python
from yellowbrick.regressor import PredictionError
visualizer = PredictionError(lm)
visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show();
```

```python
from yellowbrick.regressor import ResidualsPlot
visualizer = ResidualsPlot(lm)

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.show();
```

In this graph, it is desirable to distribute the residual values equally above and below the zero value, without forming a pattern, so that the error sums are zero.

```python
my_dict = {'Actual': y_test, 'Pred': y_test_pred}
compare = pd.DataFrame(my_dict)
```

Actual and predicted values were checked to see the difference. The model is not successful for high-priced products because there is not enough data on those prices.

There is no hyperparameter tuning in the Linear Regression model. Ridge and Lasso have been developed because overfitting will occur when there are too many features in the Linear Regression model.
